```python
# Create a line plot for one of the country's population over the years

country_name = 'Sri Lanka'
# filter and get the data for sri lanka
country_data = population[population['Country Name'] == country_name]
# Select relavant columns (years & population)
years = country_data.columns[4:] # because the years start from the 4th column
population_values = country_data[years].values[0]
# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(years, population_values, c='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {country_name} over the years')
plt.grid(True)
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# Extract population data for the year 2016
population_2016 = population['2016']

# Create a box plot to visualize outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x=population_2016, color='yellow')
plt.title('Box Plot of Population in 2016')
plt.xlabel('Population in 2016')
plt.show()
```

```python
# Identifying outliers with the IQR method
Q1 = population_2016.quantile(0.25)
Q3 = population_2016.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = population_2016[(population_2016 < lower_bound) | (population_2016 > upper_bound)]
```

```python
# Print the countries with outlier population in 2016
print("Countries with outlier population in 2016:")
for country in population[population_2016.isin(outliers)]['Country Name']:
    print(country)
```

```python
# Create a scatter plot to visualize the relationship between population and outliers in 2016
# Another method to find outliers
plt.figure(figsize=(12, 6))
plt.scatter(population['Country Name'], population_2016, color='cyan' )
plt.title('Scatter Plot of Population in 2016')
plt.xlabel('Country Name')
plt.ylabel('Population in 2016')
plt.xticks(rotation=90)
plt.show()
```

```python
# create a list of all non-country names
non_country=['Arab World', 'East Asia & Pacific (excluding high income)', 'Early-demographic dividend', 'East Asia & Pacific', 'Europe & Central Asia (excluding high income)', 'Europe & Central Asia', 'Euro area', 'European Union', 'Fragile and conflict affected situations', 'High income', 'Heavily indebted poor countries (HIPC)', 'IBRD only', 'IDA & IBRD total', 'IDA total', 'IDA blend', 'Indonesia', 'IDA only', 'Latin America & Caribbean (excluding high income)', 'Latin America & Caribbean', 'Least developed countries: UN classification', 'Low income', 'Lower middle income', 'Low & middle income', 'Late-demographic dividend', 'Middle East & North Africa', 'Middle income', 'Middle East & North Africa (excluding high income)', 'North America', 'OECD members',  'Pre-demographic dividend', 'Post-demographic dividend', 'South Asia', 'Sub-Saharan Africa (excluding high income)', 'Sub-Saharan Africa', 'East Asia & Pacific (IDA & IBRD countries)', 'Europe & Central Asia (IDA & IBRD countries)', 'Latin America & the Caribbean (IDA & IBRD countries)', 'Middle East & North Africa (IDA & IBRD countries)', 'South Asia (IDA & IBRD)', 'Sub-Saharan Africa (IDA & IBRD countries)', 'Upper middle income', 'World']
```

```python
# remove rows which are in the non_country category
filtered_population = population[~population['Country Name'].isin(non_country)]
```

```python
# Extract the 2016 data from the filtered dataset
filtered_population_2016 = filtered_population['2016']
```

```python
# Plot a box plot for this
plt.figure(figsize=(12, 6))
sns.boxplot(x=filtered_population_2016, color='purple')
plt.title('Box Plot of Population in 2016')
plt.xlabel('Population in 2016')
plt.show()
```

```python
# Identify outliers using the IQR method (after removing non-country names)
Q1_filtered = filtered_population_2016.quantile(0.25)
Q3_filtered = filtered_population_2016.quantile(0.75)
IQR_filtered = Q3_filtered - Q1_filtered
lower_bound_filtered = Q1_filtered - 1.5 * IQR_filtered
upper_bound_filtered = Q3_filtered + 1.5 * IQR_filtered

outliers_filtered = filtered_population_2016[(filtered_population_2016 < lower_bound_filtered) | (filtered_population_2016 > upper_bound_filtered)]
```

```python
# Print the countries with outlier population in 2016 (after removing non-country names)
print("Countries with outlier population in 2016 (After Removing Non-Country Names):")
for country in filtered_population[filtered_population_2016.isin(outliers_filtered)]['Country Name']:
    print(country)
```

```python
# Create a scatter plot to visualize the relationship between population and outliers in 2016 (after removing non-country names)
plt.figure(figsize=(12, 6))
plt.scatter(filtered_population['Country Name'], filtered_population_2016, c='green')
plt.title('Scatter Plot of Population in 2016 (After Removing Non-Country Names)')
plt.xlabel('Country Name')
plt.ylabel('Population in 2016')
plt.xticks(rotation=90)
plt.show()
```
