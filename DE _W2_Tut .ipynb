```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read CSV files
df_BP = pd.read_csv('/content/BUSINESS-PROCESS.csv')
df_TD = pd.read_csv('/content/timedim.csv')
df_PR = pd.read_csv('/content/phonerate.csv')
df_LO = pd.read_csv('/content/location.csv')
```

```python
# connect to database
conn = sqlite3.connect("DEdb")
cur = conn.cursor()
```

```python
# load dataframes into the database As tables with names BP,DT,PR,LO
df_BP.to_sql('BP', conn, if_exists='replace', index=False)
df_TD.to_sql('DT', conn, if_exists='replace', index=False)
df_PR.to_sql('PR', conn, if_exists='replace', index=False)
df_LO.to_sql('Lo', conn, if_exists='replace', index=False)
```

```python
# Check if data inserted to tables
BP = pd.read_sql_query("SELECT * FROM BP", conn)
BP
```
